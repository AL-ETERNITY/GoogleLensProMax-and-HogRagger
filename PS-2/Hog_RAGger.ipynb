```python
import json

with open('corpus.json', 'r') as f:
    corpus_data = json.load(f)

documents = []
for article in corpus_data:
    document = {
        'title': article['title'],
        'author': article['author'],
        'source': article['source'],
        'published_at': article['published_at'],
        'category': article['category'],
        'url': article['url'],
        'content': article['body']
    }
    documents.append(document)

print(f"Loaded {len(documents)} documents.")
```

```python


from sentence_transformers import SentenceTransformer
import chromadb
import numpy as np

model = SentenceTransformer('all-MiniLM-L6-v2')

# for document in documents:
#     document['embedding'] = model.encode(document['content'])

with open('embeddings.json', 'r') as f:
    embedding_data = json.load(f)

for i, doc in enumerate(documents):
    doc['embedding'] = np.array(embedding_data[i]['embedding'])

chroma_client = chromadb.Client()

collection = chroma_client.get_or_create_collection("documents")

for i, doc in enumerate(documents):
    print(f"Adding document {i+1}/{len(documents)}")
    metadata = {'title': doc['title'], 'url': doc['url'],'author': doc['author'],'source': doc['source'], 'category':doc['category'], 'published_at': doc['published_at']}

    for key in metadata:
        if metadata[key] is None:
            metadata[key] = ""
    collection.add(
        ids=[str(i)],
        documents=[doc['content']],
        metadatas=[metadata],
        embeddings=[doc['embedding']]
    )

```

```python
from sentence_transformers import SentenceTransformer
import chromadb
import numpy as np


# !!! Load the SAME SentenceTransformer model used for generating the stored embeddings.
# IF the 'all-MiniLM-L6-v2' model was not used to create 'embeddings.json',
# replace it with the correct model name
model = SentenceTransformer('all-MiniLM-L6-v2')

chroma_client = chromadb.Client()
collection = chroma_client.get_or_create_collection("documents")

def search_documents(query, top_k=5):

    query_embedding = model.encode(query)

    results = collection.query(
        query_embeddings=[query_embedding],
        n_results=4
    )

    return results




query = input("Enter your query: ")
result = search_documents(query)


result
```
